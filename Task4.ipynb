##       Task # 4:  Predicting House Prices using Boston Housing dataset

## Importing libraries and Basic data wranggling

```python
import numpy as np
import pandas as pd
import sklearn.preprocessing 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
data=pd.read_csv('Boston.csv')
```

```python
df=data.copy()
```

```python
df.shape
```

```python
df.head()
```

```python
df.describe(include="all")
```

```python
df.isnull().sum()
```

###  Preprocessing of categorical variables

```python
df_OneHot = pd.get_dummies(df, columns=['Residential Proportion','Charles River'])
```

```python
df=df_OneHot.copy()
```

```python
df.head()
```

## Creating independent and dependent variables

```python
X = df.drop(['median home price'], axis=1)
Y = df[['median home price']]
print(X.shape)
print(Y.shape)
```

## Normalization

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
x_scaled_minmax = min_max_scaler.fit_transform(X)
X = pd.DataFrame(x_scaled_minmax, columns = X.columns)
X.head()
```

## Train test data split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest =   train_test_split(X, Y, test_size=0.30, random_state=25,shuffle=True)
print(xtrain.shape, ytrain.shape)
print(xtest.shape, ytest.shape)
```

## Linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
```

## Prediction using traing dataset

```python
lr_predict_train = lr.predict(xtrain)
lr_predict_train[:5]
```

## Prediction using test dataset

```python
lr_predict = lr.predict(xtest)
lr_predict[:5]
```

## RMSE and R2 score

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# model evaluation for training set
y_train_predict = lr.predict(xtrain)
rmse = (np.sqrt(mean_squared_error(ytrain, y_train_predict)))
r2 = r2_score(ytrain, y_train_predict)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

# model evaluation for testing set
y_test_predict = lr.predict(xtest)
rmse = (np.sqrt(mean_squared_error(ytest, y_test_predict)))
r2 = r2_score(ytest, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```
